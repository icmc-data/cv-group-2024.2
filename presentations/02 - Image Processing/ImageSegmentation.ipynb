### Segmentação por Limiarização - Thresholding

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Carregar a imagem em escala de cinza
image = cv2.imread('/content/image.jpg', 0)

# Aplicar a limiarização global
ret, thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Exibir a imagem original e a segmentada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagem Original')

plt.subplot(1, 2, 2)
plt.imshow(thresh, cmap='gray')
plt.title('Imagem Segmentada por Limiarização')

plt.show()
```

# Segmentação por Agrupamento - Clustering


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Carregar a imagem em RGB
image = cv2.imread('/content/image.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# roi = image[1000:2500, 2500: 4500]

# Converter a imagem para um vetor de pixels
pixel_values = image.reshape((-1, 3))
pixel_values = np.float32(pixel_values)

# Critérios de parada e número de clusters
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
k = 3  # Número de clusters

# Aplicar o K-Means
_, labels, centers = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Converter os centros para inteiros (cores)
centers = np.uint8(centers)

# Mapear os rótulos para os pixels da imagem segmentada
segmented_image = centers[labels.flatten()]
segmented_image = segmented_image.reshape(image.shape)

# Exibir a imagem original e a segmentada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Imagem Original')

plt.subplot(1, 2, 2)
plt.imshow(segmented_image)
plt.title('Imagem Segmentada por agrupamento')

plt.show()
```

# segmentação por Cor

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Carregar a imagem em RGB
imgRGB = cv2.imread("/content/image.png")

# Converter a imagem para o espaço de cor HSV
imgHSV = cv2.cvtColor(imgRGB, cv2.COLOR_BGR2HSV)

# Definir os limites para segmentação de cor
tomClaro = np.array([40, 100, 100])
tomEscuro = np.array([80, 255, 255])

# Criar a máscara para segmentação da imagem dentro do intervalo de cores
imgSegmentada = cv2.inRange(imgHSV, tomClaro, tomEscuro)

# Mostrar a imagem original e a imagem segmentada
plt.figure(figsize=(10, 5))

imgRGB = cv2.cvtColor(imgRGB, cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.imshow(imgRGB)
plt.title('Imagem Original')

plt.subplot(1, 2, 2)
plt.imshow(imgSegmentada)
plt.title('Imagem Segmentada por Cor')

plt.show()
```
